```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子以确保可重复性
np.random.seed(0)

# 定义sin函数
def true_function(x):
    return np.sin(x)

# 生成x值
x = np.linspace(-5, 5, 1000)

# 生成带有高斯噪声的样本数据
def generate_data_points(x, num_points, noise_std):
    y_true = true_function(x)
    noise = np.random.normal(0, noise_std, num_points)
    y_noisy = y_true + noise
    return y_noisy

# 创建三个不同大小的数据集
S_x = np.linspace(-5, 5, 10)
S_y = generate_data_points(S_x, 10, 0.5)

M_x = np.linspace(-5, 5, 30)
M_y = generate_data_points(M_x, 30, 0.5)

L_x = np.linspace(-5, 5, 60)
L_y = generate_data_points(L_x, 60, 0.5)

# 定义多项式基函数
def polynomial_basis(x, K):
    basis = np.zeros((len(x), K + 1))
    for i in range(K + 1):
        basis[:, i] = x**i
    return basis

# 实现贝叶斯线性回归
def bayesian_regression(X, y, alpha, beta):
    XTX = np.dot(X.T, X)
    XTy = np.dot(X.T, y)
    cov_inv = alpha * np.identity(XTX.shape[0]) + beta * XTX
    cov = np.linalg.inv(cov_inv)
    mean = beta * np.dot(cov, XTy)

    print(y)
    print(mean)
    return mean, cov

# 绘制多项式拟合结果
def plot_polynomial_fit(x, y, means, covs, K, num_samples=20):
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, label='数据点', color='blue')
    plt.plot(x, true_function(x), label='sin函数', color='green')
    
    X_poly = polynomial_basis(x, K)
    
    for i in range(num_samples):
        sample_weights = np.random.multivariate_normal(means, covs)
        y_pred = np.dot(X_poly, sample_weights)
        plt.plot(x, y_pred, label=f'样本{i+1}', alpha=0.5)
    
    plt.title(f'K={K} 拟合结果')
    plt.legend()
    plt.ylim(-2, 2)
    plt.show()

# 贝叶斯线性回归参数
alpha = 0.01
beta = 0.5

# 对S、M、L三个数据集分别进行拟合和绘制
K_values = [1, 3, 7]

for K in K_values:
    X_S = polynomial_basis(S_x, K)
    mean_S, cov_S = bayesian_regression(X_S, S_y, alpha, beta)
    plot_polynomial_fit(S_x, S_y, mean_S, cov_S, K, num_samples=20)

    X_M = polynomial_basis(M_x, K)
    mean_M, cov_M = bayesian_regression(X_M, M_y, alpha, beta)
    plot_polynomial_fit(M_x, M_y, mean_M, cov_M, K, num_samples=20)

    X_L = polynomial_basis(L_x, K)
    mean_L, cov_L = bayesian_regression(X_L, L_y, alpha, beta)
    plot_polynomial_fit(L_x, L_y, mean_L, cov_L, K, num_samples=20)
```
